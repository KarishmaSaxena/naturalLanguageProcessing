Use stemming  when speed matters (search engines, indexing)

Use lemmatization  when meaning matters (chatbots, sentiment analysis)

Lemmatization is used to reduce words to their meaningful base form by considering grammar and context, which improves accuracy in NLP tasks.

Lemmatization is used to overcome the limitations of stemming.

the ouput we get after lemmatization is called "lemma"

stemming - Cuts words mechanically

Ignores context and grammar

Produces non-meaningful words

lemmatizing - Uses dictionary + grammar rules

Considers part of speech (POS)

Returns valid, meaningful words

 WordNet Lemmatization?

 Is a lemmatization technique that uses Dictionary Lookup (WordNet), a large lexical database of English.

It converts words into their meaningful base form (lemma) using dictionary rules.

Implemented using NLTKs WordNetLemmatizer.

Why WordNet Lemmatization is Used ->>>
Produces valid dictionary words
Reduces over-stemming errors
Considers part of speech (POS)
Improves semantic understanding in NLP tasks

How WordNet Lemmatization Works ->>>
Takes a word as input
Checks WordNet dictionary
Uses POS tag (if provided)
Returns the correct lemma


```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('car', pos='n') ,lemmatizer.lemmatize('cars', pos='n') ,  lemmatizer.lemmatize('carred', pos='v') ,lemmatizer.lemmatize('are', pos='v')  , lemmatizer.lemmatize('is', pos='v')  , lemmatizer.lemmatize('was', pos='v') , lemmatizer.lemmatize('were', pos='v')  , lemmatizer.lemmatize('going', pos='v') ,lemmatizer.lemmatize('moving', pos='v')  , lemmatizer.lemmatize('move', pos='v')  , lemmatizer.lemmatize('movement', pos='n')
```

types - lemmatization can be dictionary-based, rule-based, POS-aware, or ML-based, depending on how linguistic knowledge is applied.

```python
#Dictionary-Based Lemmatization (WordNet  NLTK)
words = ['cars', 'was', 'better', 'going']
for w in words:
    print(w, "", lemmatizer.lemmatize(w))
```

```python
#POS-Aware Lemmatization (WordNet + POS)
print(lemmatizer.lemmatize('cars', pos='n'))
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("cars were moving better than children")
for token in doc:
   print(token.text, "", token.lemma_)
```
