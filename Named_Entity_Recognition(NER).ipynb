```python
sentence = "Elon Musk bought a spaceship from NASA and flew it to Mars yesterday. Harry Potter had a meeting with Barack Obama at Hogwarts. Starbucks ran out of coffee in New York, and everyone panicked. Pikachu chased me while I was walking in Central Park. Jeff Bezos ordered a pizza from Domino’s on his way to the Moon."
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
words = nltk.word_tokenize(sentence)
words
```

```python
nltk.download('averaged_perceptron_tagger_eng')
tag_elements = nltk.pos_tag(words)
```
