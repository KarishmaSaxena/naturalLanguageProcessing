Porter Stemmer

Most widely used algorithm

Uses a set of rules to remove suffixes

Moderate speed and accuracy

Example:

running  run

relational  relat

```python
from nltk.stem import PorterStemmer

porter_stemmer = PorterStemmer()

words = ["running", "jumps", "happily", "running", "happily"]

stemmed_words = [porter_stemmer.stem(word) for word in words]

print("Original words:", words)
print("Stemmed words:", stemmed_words)
```

Snowball Stemmer (Porter2)

Improved version of Porter Stemmer

More accurate and consistent

Supports multiple languages

Example:

better  better



```python
#snowball stemming
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer(language='english')

words_to_stem = ['running', 'jumped', 'happily', 'quickly', 'foxes']

stemmed_words = [stemmer.stem(word) for word in words_to_stem]

print("Original words:", words_to_stem)
print("Stemmed words:", stemmed_words)
```

```python
#porter
porter_stemmer.stem('fairly'), porter_stemmer.stem('sportingly')
```

```python
# snwoball
stemmer.stem('fairly'), stemmer.stem('sportingly')
```

```python
#regexstemmer
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|ables$' , min=4)
```

```python
reg_stemmer.stem('eating') , reg_stemmer.stem('studying')
```

 Lancaster Stemmer

Very aggressive stemming

Removes more characters

Fast but less accurate

Example:

maximum  maxim

caring  car

```python
from nltk.stem import LancasterStemmer
lancaster_stemmer = LancasterStemmer()
```

```python
lancaster_stemmer.stem('eating') , lancaster_stemmer.stem('caring'), lancaster_stemmer.stem('studying') , lancaster_stemmer.stem('fairly') , lancaster_stemmer.stem('sportingly') , lancaster_stemmer.stem('running')
```
