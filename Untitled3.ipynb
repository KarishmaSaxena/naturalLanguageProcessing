```python
# Load dataset
dataset = pd.read_csv('/content/Reviews.csv', nrows=250000)
```

```python
print("Dataset shape:", dataset.shape)
print("Sample review text:\n", dataset['Text'][0])
```

```python
# EDA: Count of reviews by score
dataset['Score'].value_counts().sort_index().plot(
    kind='bar', title='Count of reviews by stars', xlabel='Score', ylabel='Count', figsize=(10,5)
)
plt.show()
```

```python


# NLTK Basic Example
example = dataset['Text'][0]

# Tokenize words and sentences
words = word_tokenize(example)
sentences = sent_tokenize(example)

print("Words:", words)
print("Sentences:", sentences)

# Stemming & Lemmatization
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print("Stemmed:", ps.stem(example))
print("Lemmatized:", lemmatizer.lemmatize(example))
```

```python
# POS tagging
pos_tags = pos_tag(words)
print("POS Tags:", pos_tags)
```

```python
# Named Entity Recognition
named_entities = ne_chunk(pos_tags)
print("Named Entities:", named_entities)
```

```python
# VADER Sentiment Analysis
sentiment_analyzer = SentimentIntensityAnalyzer()
print("VADER example score:", sentiment_analyzer.polarity_scores(example))
```

```python
# Example sentences
print("Happy:", sentiment_analyzer.polarity_scores("I am so happy"))
print("Sad:", sentiment_analyzer.polarity_scores("I am so sad"))
```

```python
# VADER Scores for first 500 reviews
df = dataset.head(500)
res = {}
for index, row in tqdm(df.iterrows(), total=len(df)):
    text = row['Text']
    myid = row['Id']
    res[myid] = sentiment_analyzer.polarity_scores(text)

vaders = pd.DataFrame(res).T.reset_index().rename(columns={'index': 'Id'})
vaders = vaders.merge(df, how='left')
```

```python
# Plot compound scores by review score
sns.barplot(data=vaders, x='Score', y='compound')
plt.xlabel('Score')
plt.ylabel('Compound Score')
plt.title('Compound Score by Review Score')
plt.show()
```

```python
# Plot positive, negative, neutral by review score
fig, axs = plt.subplots(1, 3, figsize=(15,5))
sns.barplot(data=vaders, x='Score', y='pos', ax=axs[0], color='blue')
sns.barplot(data=vaders, x='Score', y='neg', ax=axs[1], color='red')
sns.barplot(data=vaders, x='Score', y='neu', ax=axs[2], color='green')

axs[0].set_title('Positive')
axs[1].set_title('Negative')
axs[2].set_title('Neutral')
plt.show()
```

```python
# RoBERTa Sentiment Analysis
MODEL = "cardiffnlp/roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)

def polarity_scores_roberta(text):
    encoded_text = tokenizer(text, return_tensors='pt')
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    return {
        'roberta_neg': scores[0],
        'roberta_neu': scores[1],
        'roberta_pos': scores[2]
    }
```

```python
# Merge VADER and RoBERTa for first 500 reviews
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    try:
        text = row['Text']
        myid = row['Id']
        vaders_result = sentiment_analyzer.polarity_scores(text)
        roberta_result = polarity_scores_roberta(text)
        res[myid] = {**vaders_result, **roberta_result}
    except RuntimeError:
        print(f'Error for Id {myid}')

results_df = pd.DataFrame(res).T.reset_index().rename(columns={'index':'Id'})
results_df = results_df.merge(df, how='left')
```

```python
# Compare VADER vs RoBERTa
sns.pairplot(
    data=results_df,
    vars=['neg','neu','pos','roberta_neg','roberta_neu','roberta_pos'],
    hue='Score', palette='tab10'
)
plt.show()
```

```python
# Review Examples
print("Most positive 1-star review by RoBERTa:")
print(results_df.query('Score == 1').sort_values('roberta_pos', ascending=False)['Text'].values[0])

print("Most negative 5-star review by VADER:")
print(results_df.query('Score == 5').sort_values('neg', ascending=False)['Text'].values[0])
```
